```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```python
# Load the data
data = pd.read_csv('dataset.csv')
```

```python
# Convert categorical features to numerical if necessary
data['posture'] = data['posture'].astype('category').cat.codes
data['muscle'] = data['muscle'].astype('category').cat.codes
data['exercise'] = data['exercise'].astype('category').cat.codes
```

```python
# Define features and target
X = data[['body_weight', 'muscle', 'weight_lifted', 'exercise', 'EMG_180', 'EMG_90', 'EMG_20']]
y = data['posture']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Define features and target
X = data[['body_weight', 'muscle', 'weight_lifted', 'exercise', 'EMG_180', 'EMG_90', 'EMG_20']]
y = data['posture']
```

```python
# Feature scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train the Naive Bayes model
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
# Predict the test set results
y_pred = nb.predict(X_test)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
# Print classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
plt.figure(figsize=(10,7))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.colorbar()
tick_marks = np.arange(len(np.unique(y)))
plt.xticks(tick_marks, np.unique(y), rotation=45)
plt.yticks(tick_marks, np.unique(y))

# Annotate the confusion matrix
thresh = cm.max() / 2.
for i, j in np.ndindex(cm.shape):
    plt.text(j, i, format(cm[i, j], 'd'),
             horizontalalignment="center",
             color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```
